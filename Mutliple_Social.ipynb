```python
import pandas as pd
import numpy as np

df = pd.read_excel("multiple_social.xlsx")

df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.isnull().sum()
```

```python
# numeric columns
for col in df.select_dtypes(include=np.number):
    df[col] = df[col].fillna(df[col].mean())

# categorical columns
for col in df.select_dtypes(include='object'):
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip().str.lower()
```

```python
df.dtypes
```

```python
import numpy as np

# convert all numeric columns to float
for col in df.select_dtypes(include='number'):
    df[col] = df[col].astype(float)
```

```python
for col in df.select_dtypes(include='number'):
    idx = np.random.choice(df.index, size=int(0.05 * len(df)), replace=False)
    
    df.loc[idx, col] = df[col].mean() + (15 * df[col].std())
```

```python
df_before = df.copy()
```

```python
for col in df.select_dtypes(include='number'):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    count = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col} → Outliers BEFORE: {count}")
```

```python
import matplotlib.pyplot as plt

for col in df_before.select_dtypes(include='number'):
    plt.figure()
    plt.boxplot(df_before[col])
    plt.title(f"Before Outlier Treatment: {col}")
    plt.xlabel(col)
    plt.ylabel("Values")
    plt.show()
```

```python
for col in df.select_dtypes(include='number'):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
```

```python
for col in df.select_dtypes(include='number'):
    plt.figure()
    plt.boxplot(df[col])
    plt.title(f"After Outlier Treatment: {col}")
    plt.xlabel(col)
    plt.ylabel("Values")
    plt.show()
```

```python
# Engagement rate
df['engagement_rate'] = df['total_engagement'] / df['followers']

# Content performance
df['performance_score'] = df['likes'] + df['shares'] + df['comments']

# Viral efficiency
df['viral_efficiency'] = df['views'] / df['followers']
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
```

```python
df.to_csv("clean_social_evolution.csv", index=False)
```
